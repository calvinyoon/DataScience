# Manipulating Data

```python
car_sales["Price"] = car_sales["Price"].str.replace('[\$\,\.]', "").astype(int) // 100
```

```python
car_sales
```

```python
car_sales["Make"] = car_sales["Make"].str.lower()
car_sales
```

```python
car_sales_missing = pd.read_csv("https://raw.githubusercontent.com/mrdbourke/zero-to-mastery-ml/master/data/car-sales-missing-data.csv")
car_sales_missing
```

```python
car_sales_missing["Odometer"] = car_sales_missing["Odometer"].fillna(car_sales_missing["Odometer"].mean())
car_sales_missing
```

```python
car_sales_missing_dropped = car_sales_missing.dropna()
car_sales_missing_dropped
```

```python
#create new column
seats_column = pd.Series([5, 5, 5, 5, 5])
car_sales["Seats"] = seats_column
car_sales
```

```python
car_sales["Seats"].fillna(5, inplace=True)
car_sales
```

```python
#column from python list
fuel_economy = [7.5, 9.5, 5.0, 9.6, 8.7, 4.7, 7.6, 8.7, 3.0, 4.5]
car_sales["Fuel per 100KM"] = fuel_economy
car_sales
```

```python
#do math and add column
car_sales["Total fuel used (L)"] = car_sales["Odometer (KM)"]/100 * car_sales["Fuel per 100KM"]
car_sales
```

```python
car_sales["Number of wheels"] = 4
car_sales
```

```python
car_sales["Passed road safety"] = True
car_sales
```

```python
car_sales = car_sales.drop("Total fuel used (L)", axis=1) #drop column
car_sales
```

```python
car_sales_shuffled = car_sales.sample(frac=0.5) #sample 50% of data
car_sales_shuffled
```

```python
#reset index of sampled table
car_sales_shuffled.reset_index(drop=True, inplace=True)
car_sales_shuffled
```

```python
car_sales
```

```python
car_sales["Odometer (KM)"] = car_sales["Odometer (KM)"].apply(lambda x: x / 1.6)  #lambda is anon func. replace x with x/1.6
car_sales
```
